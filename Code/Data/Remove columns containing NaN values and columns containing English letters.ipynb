```python
import pandas as pd
```

```python
data = pd.read_csv('nd2_para.csv')

columns_with_nan = data.columns[data.isnull().any()]  
data.drop(columns=columns_with_nan, inplace=True) 
print(data)
```

```python
data.to_csv('cleaned_nd2.csv', index=False)  
```

```python
data = pd.read_csv('cleaned_nd2.csv')
columns_to_drop = []
for col in data.columns:
    if data[col].dtype == 'object': 
        if data[col].str.contains('[a-zA-Z]').any():  
            columns_to_drop.append(col)

data.drop(columns=columns_to_drop, inplace=True)

data.to_csv('cleanall_nd2.csv', index=False)  
```
